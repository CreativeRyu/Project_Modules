### **<span style="color:deepskyblue">Asteroid x Laser Collision</span>**

## **<span style="color:#fcba04">Code</span> Challenge**


+ implementiere eine Kollision zwischen den Laser- und den Asteroiden-Objekten
+ hierfür werden zwei for loops benötigt
+ wenn eine berührung statt gefunden hat, sollen sowohl das Laser- als auch das Asteroiden-Objekt entfernt werden


<div class="alert alert-block alert-success">
<b>Lösung Asteroid x Laser Collision</b>
</div>

```python
# Collisions
for laser in laser_list:
    for asteroid_tuple in asteroid_list:
        asteroid = asteroid_tuple[0]
        if laser.colliderect(asteroid):
            asteroid_list.remove(asteroid_tuple)
            laser_list.remove(laser)
```
