# **<span style="color:deepskyblue">Rectangles</span>**

+ werden für Kollisionen eingesetzt
+ machen es einfacher ein Surface präzise zu platzieren

***

Statt in der *blit()* Methode eine Position zu übergeben, kann einfach ein Rectangle verwendet werden.

##### *<span style="color:skyblue">display_surface</span>.blit(<span style="color:hotpink">surface</span>, rect)*

In der Regel wird immer ein <span style="color:hotpink">Surface</span> und ein Rectangle erstellt.\
Das <span style="color:hotpink">Surface</span> speichert die Bilddaten, das Rectangle speichert Position und Bewegung.

Die Bewegung von Objekten im Spiel wird also nicht über ein <span style="color:hotpink">Surface</span> sondern über ein Rectangle erstellt.


***

```python
import pygame
import sys
path = "Game Dev Projects/01 Space Shooter/"

pygame.init()
clock = pygame.time.Clock()
WINDOW_WIDTH, WINDOW_HEIGHT = 1280, 720
display = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))

ship_surface = pygame.image.load(path+"images/player/SmartSpaceShip.png").convert_alpha()
# Ship Y Koordinate ausbauen und ein Rectangle hier rein
ship_rectangle = ship_surface.get_rect(center = (640,360)) # hier können wir das center für eine präzisere Platzierung verwenden

background_surface = pygame.image.load(path+"./images/background/background1.png").convert()
font = pygame.font.Font(path+"./fonts/subatomic.ttf", 50)
text_surface = font.render("Space Boi", True, "White")

# GAME LOOP
while True:
    # 1. Player Inputs 
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit() 
            sys.exit()
    
    #Limiting Frame Rate
    clock.tick(60)
    
    # 2. Updates
    display.fill((0, 0, 0))
    display.blit(background_surface, (0, 0))
    display.blit(ship_surface, ship_rectangle) # hier unten dann das Rectangle einbauen
    display.blit(text_surface, (500, 150))
    
    # 3. Show Frame to Player -> update display_surface
    pygame.display.update()
```

## **<span style="color:#fcba04">Code</span> Challenge**

Erstelle ein Rectangle vom Titeltext des Spiels und\
platziere dieses ebenfalls oben mittig im <span style="color:skyblue">Display Surface</span>.


<div class="alert alert-block alert-success">
<b>Lösung Text Rectangle</b>
</div>

```python
import pygame
import sys
path = "Game Dev Projects/01 Space Shooter/"

pygame.init()
clock = pygame.time.Clock()
WINDOW_WIDTH, WINDOW_HEIGHT = 1280, 720
display = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))

ship_surface = pygame.image.load(path+"images/player/SmartSpaceShip.png").convert_alpha()
# Ship Y Koordinate ausbauen und ein Rectangle hier rein
ship_rectangle = ship_surface.get_rect(center = (WINDOW_WIDTH / 2, WINDOW_HEIGHT / 2)) 
# hier können wir das center für eine präzisere Platzierung verwenden
background_surface = pygame.image.load(path+"./images/background/background1.png").convert()
font = pygame.font.Font(path+"./fonts/subatomic.ttf", 50)
titel_surface = font.render("Space Boi", True, "White")
titel_rectangle = titel_surface.get_rect(center=(WINDOW_WIDTH / 2, 150))

# GAME LOOP
while True:
    # 1. Player Inputs 
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit() 
            sys.exit()
    
    #Limiting Frame Rate
    clock.tick(60)
    
    # 2. Updates
    display.fill((0, 0, 0))
    display.blit(background_surface, (0, 0))
    display.blit(ship_surface, ship_rectangle) # hier unten dann das Rectangle einbauen
    display.blit(titel_surface, titel_rectangle)
    
    # 3. Show Frame to Player -> update display_surface
    pygame.display.update()
```

## **<span style="color:deepskyblue">Rectangle bewegen</span>**

+ top, bottom, y und centery vertikal
+ left, right, x und centerx horizontal

```python
# 2. Updates
display.fill((0, 0, 0))
display.blit(background_surface, (0, 0))
ship_rectangle.top -= 3
display.blit(ship_surface, ship_rectangle)
display.blit(titel_surface, titel_rectangle)
```
